```python
import torch 
import torch.nn as nn
from sklearn.preprocessing import LabelEncoder
```

```python
from ucimlrepo import fetch_ucirepo 
  
# fetch dataset 
iris = fetch_ucirepo(id=53) 
  
# data (as pandas dataframes) 
X = iris.data.features 
y = iris.data.targets 
```

```python
X.shape, y.shape
```

```python
X_tensor = torch.tensor(X.to_numpy(),dtype = torch.float32,requires_grad=True)
```

```python
y_boolean = LabelEncoder().fit_transform(y)
```

```python
y_bol_ten = torch.tensor(y_boolean,dtype = torch.float32,requires_grad=True).long()
```

```python
y_tensor = nn.functional.one_hot(y_bol_ten,num_classes=4)
```

```python
y_tensor #print(#y_tensor) prints full tensors
```

```python
y_tensor = y_tensor.type(torch.float32)
```

```python
y_tensor.shape
```

```python
model = nn.Sequential(nn.Linear(4, 2),
                      nn.Linear(2, 4),
                      nn.ReLU()
                     )
```

```python
model
```

```python
y_prediction = model(X_tensor)
```

```python
y_prediction
```

```python
y_prediction.shape == y_tensor.shape
```

```python
loss_object = nn.CrossEntropyLoss()
```

```python
loss = loss_object(y_prediction,y_tensor)
```

```python
loss
```

```python
loss.backward() #Computes the gradient of current tensor wrt graph leaves.
```

```python
wieght_0 = model[0].weight
```

```python
wieght_0
```

```python
wieght_0_grad = wieght_0.grad
wieght_0_grad
```

```python
bias_0 = model[0].bias
bias_0
```

```python
bias_0_grad = bias_0.grad
bias_0_grad
```

```python
lr = 0.1
bias_0 = bias_0 - lr*bias_0_grad
```

```python
bias_0
```

```python
## You cannot do manually for many layers and for each layer you have to update the many parameters
## PyTorch optimizers handle it all in just one line of code
```

```python
optimizer = torch.optim.SGD(model.parameters(),lr = 0.1)
```

```python
optimizer
```

```python
optimizer.step() #Perform a single optimization step.
```

```python
# optimizer.zero_grad() #Prevents gradient accumulation from previous step.
```

```python
loss_2 = loss_object(y_pred_2,y_tensor)
```

```python
loss_2
```

```python
loss_2.backward() ## compute new gradients
```
