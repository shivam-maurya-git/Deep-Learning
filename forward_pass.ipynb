```python
import torch
import torch.nn as nn
import pandas as pd
```

```python
df = pd.read_csv("diabetes.csv")
```

```python
df.head()
```

```python
X = df.drop("Outcome",axis=1)
```

```python
X.shape
```

```python
X_array = X.to_numpy()
```

```python
X_array
```

```python
# tensor expect data like numpy array, not additional metadata like col, indices
inputs = torch.tensor(data = X_array)
```

```python
inputs
```

```python
model = nn.Sequential(
    nn.Linear(8,3,dtype = torch.float64), # need to specify, bceause default is dtype = torch.float32
    nn.Linear(3,1,dtype = torch.float64),
    nn.Sigmoid())
```

```python
model(inputs)
```
